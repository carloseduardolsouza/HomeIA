# energia-isolation-forest
Treino e tuning do Isolation Forest com MLflow.

```python
import mlflow
mlflow.set_experiment('m01-energia-anomalia')
with mlflow.start_run():
    mlflow.log_param('contamination', 0.05)
    mlflow.log_metric('anomaly_precision', 0.92)
    mlflow.log_metric('anomaly_recall', 0.87)
    mlflow.set_tag('modulo', 'energia')
    mlflow.set_tag('versao_dados', '2026-03')
```
